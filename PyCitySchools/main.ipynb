```python
# Imports
import csv
import pandas as pd
from pathlib import Path
import numpy as np

# Find file path
school_file  = Path ("Resources\schools_complete.csv")
student_file = Path ("Resources\students_complete.csv")

# Read into data frames
school_file_df = pd.read_csv(school_file)
student_data_df = pd.read_csv(student_file)

# Merge Data into one Data Set
merged_df = pd.merge(student_data_df, school_file_df, how = "left", on = ["school_name", "school_name"])
merged_df.head()
```

```python
# Calculate the total number of unique schools

school_names = merged_df["school_name"].unique()
school_total = len(school_names)
print(school_total)
```

```python
# Calculate the total number of students

total_students = len(merged_df["student_name"])
print(total_students)
```

```python
# Calculate the total budget #Isn't calculating right

total_budget = sum(merged_df["budget"].unique())
total_budget
```

```python
# Calculate the average mean math score

average_math = merged_df["math_score"].mean()
average_math
```

```python
# Calculate the average mean reading score

average_reading = merged_df["reading_score"].mean()
average_reading
```

```python
#% passing math (the percentage of students who passed math

passing_math = merged_df[(merged_df ["math_score"] >= 70)].count()["student_name"]
passing_math_percentage = passing_math/float(total_students) * 100
passing_math_percentage

```

```python
#% passing reading (the percentage of students who passed rea#ding)

passing_reading_count = merged_df[(merged_df ["reading_score"] >= 70)].count()["student_name"]
passing_reading_percentage = passing_reading_count/float(total_students) * 100
passing_reading_percentage
```

```python
#% overall passing (the percentage of students who passed mat#h AND reading)

passing_reading_math_count = merged_df[
    (merged_df["math_score"] >= 70) & (merged_df["reading_score"]>= 70)
].count()["student_name"]
overall_passing_rate = passing_reading_math_count/float(total_students) * 100
overall_passing_rate
```

```python
# Create a high-level snapshot of the district's key metrics in a DataFrame

district_summary_df = pd.DataFrame({
    "Total Schools": [school_total],
   "Total Students": [total_students],
    "Total Budget": [total_budget],
    "Average Math Score": [round(average_math, 2)],
    "Average Reading Score": [round(average_reading, 2)],
    "% Passing Math": [round(passing_math_percentage, 2)],
    "% Passing Reading": [round(passing_reading_percentage, 2)],
    "% Overall Passing": [round(overall_passing_rate, 2)]
})
# Formatting

district_summary_df = district_summary_df.set_index("Total Schools")

# Display the DataFrame

district_summary_df
```

```python
# Use the code provided to select all of the school types

school_types = merged_df.groupby("school_name")["type"].unique()
numpy_array = school_types.to_numpy()
school_type_array = np.concatenate(numpy_array)
print(school_type_array)
```

```python
# Calculate the total student count per school

students_per_school = merged_df.groupby("school_name")["size"].mean()
students_per_school
```

```python
# Calculate the total school budget and per capita spending per school

per_school_budget = merged_df.groupby("school_name")["budget"].mean()
per_school_capita = per_school_budget/merged_df.groupby("school_name")["size"].mean()
print(per_school_budget)
print(per_school_capita)
```

```python
# Calculate the average test scores per school

per_school_math = merged_df.groupby("school_name")["math_score"].agg('mean')
print(per_school_math)

per_school_reading = merged_df.groupby("school_name")["reading_score"].agg('mean')
print(per_school_reading)
```

```python
# Calculate the number of students per school with math scores of 70 or higher

students_passing_math = merged_df[(merged_df ["math_score"] >= 70)].count()["student_name"]
school_students_passing_math = merged_df.groupby('school_name')['math_score'].agg(lambda x: (x >= 70).sum()).astype(float)
print(school_students_passing_math)
```

```python
# Calculate the number of students per school with reading scores of 70 or higher

students_passing_reading = merged_df[(merged_df ["reading_score"] >= 70)].count()["student_name"]
school_students_passing_reading = merged_df.groupby('school_name')['reading_score'].agg(lambda x: (x >= 70).sum())
print(school_students_passing_reading)
```

```python
# Use the provided code to calculate the number of students per school that passed both math and reading with scores of 70 or higher

students_passing_math_and_reading = merged_df[
    (merged_df["reading_score"] >= 70) & (merged_df["math_score"] >= 70)
]
school_students_passing_math_and_reading = students_passing_math_and_reading.groupby(["school_name"]).size()
print(school_students_passing_math_and_reading)
```

```python
# Use the provided code to calculate the passing rates

percent_school_passing_math = school_students_passing_math / students_per_school * 100
percent_school_passing_reading = school_students_passing_reading / students_per_school * 100
overall_passing_rate = school_students_passing_math_and_reading / students_per_school * 100
print(percent_school_passing_math)
print(percent_school_passing_reading)
print(overall_passing_rate)
```

```python
# Create a DataFrame called `per_school_summary` with columns for the calculations above.

per_school_summary = pd.DataFrame({
    "School Type" : school_type_array,
    "Total Students": students_per_school,
    "Total School Budget": per_school_budget,
    "Per Student Budget": round(per_school_capita,2),
    "Average Math Score": round(per_school_math, 2),
    "Average Reading Score": round(per_school_reading, 2),
    "% Passing Math": round(percent_school_passing_math, 2),
    "% Passing Reading": round(percent_school_passing_reading, 2),
    "% Overall Passing": round(overall_passing_rate, 2)
})
per_school_summary["Total School Budget"] = per_school_summary["Total School Budget"].map("${:,.2f}".format)
per_school_summary["Per Student Budget"] = per_school_summary["Per Student Budget"].map("${:,.2f}".format)

# Display the DataFrame
per_school_summary
```

```python
# Sort the schools by `% Overall Passing` in descending order and display the top 5 rows.
top_schools = per_school_summary["% Overall Passing"].sort_values(ascending= False)
top_schools.head(5)
```

```python
# Sort the schools by `% Overall Passing` in ascending order and display the top 5 rows.
bottom_schools = per_school_summary["% Overall Passing"].sort_values(ascending=True)
bottom_schools.head(5)
```

```python
# Establish the bins 
bins = [0.0, 585.00, 630.00, 645.00, 680.00]
bin_labels = ["<585", "585-630", "630-645", "645-680"]
```

```python
# Create a copy of the school summary since it has the "Per Student Budget" 
school_spending_df = per_school_summary.copy()

per_school_summary["Per Student Budget"] = per_school_summary["Per Student Budget"].str.replace('[\$,]', '', regex=True).astype(float)
```

```python
# Use `pd.cut` to categorize spending based on the bins.
school_spending_df["Spending Ranges (Per Student)"] = pd.cut(per_school_summary["Per Student Budget"],
                                                             bins, labels=bin_labels, include_lowest=True)

school_spending_df
```

```python
#  Calculate averages for the desired columns. 
spending_math_scores = school_spending_df.groupby(["Spending Ranges (Per Student)"])["Average Math Score"].mean()
spending_reading_scores = school_spending_df.groupby(["Spending Ranges (Per Student)"])["Average Reading Score"].mean()
spending_passing_math = school_spending_df.groupby(["Spending Ranges (Per Student)"])["% Passing Math"].mean()
spending_passing_reading = school_spending_df.groupby(["Spending Ranges (Per Student)"])["% Passing Reading"].mean()
overall_passing_spending = school_spending_df.groupby(["Spending Ranges (Per Student)"])["% Overall Passing"].mean()
print(spending_math_scores,spending_reading_scores,spending_passing_math,spending_passing_reading,overall_passing_spending)
```

```python
# Assemble into DataFrame
spending_summary = pd.DataFrame ({
    "Avg Math Scores": spending_math_scores,
    "Avg Reading Scores": spending_reading_scores,
    "Passing Math Scores": spending_passing_math,
    "Passing Math Scores": spending_passing_reading,
    "Overall Passing": overall_passing_spending
})

# Display results
spending_summary
```

```python
# Establish the bins.
size_bins = [0, 1000, 2000, 5000]
labels = ["Small (<1000)", "Medium (1000-2000)", "Large (2000-5000)"]
```

```python
# Categorize the spending based on the bins
# Use `pd.cut` on the "Total Students" column of the `per_school_summary` DataFrame.

per_school_summary["School Size"] = pd.cut(per_school_summary["Total Students"],
                                           size_bins, labels = labels, include_lowest = True)      
per_school_summary
```

```python
# Calculate averages for the desired columns. 
size_math_scores = per_school_summary.groupby(["School Size"])["Average Math Score"].mean()
size_reading_scores = per_school_summary.groupby(["School Size"])["Average Reading Score"].mean()
size_passing_math = per_school_summary.groupby(["School Size"])["% Passing Math"].mean()
size_passing_reading = per_school_summary.groupby(["School Size"])["% Passing Reading"].mean()
size_overall_passing = per_school_summary.groupby(["School Size"])["% Overall Passing"].mean()
```

```python
# Create a DataFrame called `size_summary` that breaks down school performance based on school size (small, medium, or large).
# Use the scores above to create a new DataFrame called `size_summary`
size_summary = pd.DataFrame({
    "Average Math Score": size_math_scores,
    "Average Reading Score":size_reading_scores,
    "% Passing Math":size_passing_math,
    "% Passing Reading":size_passing_reading,
    "% Overall Passing":size_overall_passing
})

# Display results
size_summary
```
